Descripcion: Solucion de sistema de primer orden

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funci√≥n a resolver
# dy/dt = -1/T*y+1/T*u(t)
# parametros
T = 6
def f(y,u,t):
    sal = -1/T*y+1/T*u
    return sal
# condiciones iniciales
y0 = -42

# Tiempo de simulacion
tfin = 40
h = 1
t = np.arange(0,tfin,h)

# Funcion de entrada
u = np.ones(len(t))

# variables informaticas
y = np.zeros(len(t))
y[0] = y0
for i in range(len(t)-1):
    k1 = f(y[i],u[i],t[i])
    ytemp = y[i] + h * k1
    k2 = f(ytemp,u[i+1],t[i+1])
    
    y[i+1] = y[i] + h*(k1+k2)/2
```

```python
plt.plot(t,u)
plt.plot(t,y)
plt.title('respuesta al escalon de un sistema de primer orden')
plt.xlabel('Tiempo (s)')
plt.ylabel('y')
plt.grid(visible=True)
plt.show()
```
