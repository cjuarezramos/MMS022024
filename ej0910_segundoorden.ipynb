```python
import numpy as np
import matplotlib.pyplot as plt



# sistema de segundo orden masa resorte amortiguador
# caso 1: s1 y s2 reales y diferentes
M = 5
K = 1
B = 6
x0 = 1
v0 = 0

arg = B**2/(2*M)**2 - K/M
s1 = -B/(2*M) + arg**0.5
s2 = -B/(2*M) - arg**0.5
print("s1:",s1,"y s2:",s2)
A1 = (x0*s2-v0)/(s2-s1)
A2 = (v0-x0*s1)/(s2-s1)
print("A1:",A1,"y A2:", A2)

t = np.arange(0,25,0.1)
x = A1*np.exp(s1*t)+A2*np.exp(s2*t)
plt.plot(t,x)
```
