```python
import numpy as np
import matplotlib.pyplot as plt
# definicion de parametros
c = 12.5
m = 60.1
g = 9.81
```

```python
def euler_method(f, y0, t0, t_end, h):
    # Inicializar variables
    t_values = np.arange(t0, t_end, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0

    # Aplicar el método de Euler
    for i in range(1, len(t_values)):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])

    return t_values, y_values

# Definir la función que representa la ecuación diferencial
def f(t, v):
    return g - c/m * v  # Ejemplo: dy/dt = y - t^2 + 1

# Parámetros
y0 = 0  # Valor inicial
t0 = 0    # Tiempo inicial
t_end = 20 # Tiempo final
h = 0.1   # Paso

# Llamar al método de Euler
t_values, y_values = euler_method(f, y0, t0, t_end, h)

# Graficar los resultados
plt.plot(t_values, y_values, label='Método de Euler')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solución de EDO usando el método de Euler')
plt.legend()
plt.grid()
plt.show()
```
