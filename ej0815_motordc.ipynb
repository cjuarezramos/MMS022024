```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# constantes del sistema
R = 2.27
L = 0.0047
Kb = 0.25
Km = 0.25
J =  0.00246
B = 0.003026
```

```python
# Datos de simulacion
# paso de integracion
h = 0.0001
tf = 3
m = int(tf/h)
print(m)
```

```python
i = np.zeros(m)
w = np.zeros(m)
t = np.zeros(m)
# condiciones iniciales
i0=0
w0=0
i[0]=i0
w[0]=w0
for k in np.arange(m-1):
    Vin = 1
    if t[k] < 1.5:
        TL = 0
    else:
        TL = 0.1
        
    i[k+1]=i[k]+h*(-R/L*i[k]-Kb/L*w[k]+1/L*Vin)
    w[k+1]=w[k]+h*(Km/J*i[k]-B/J*w[k]-1/J*TL)
    t[k+1]=t[k]+h

plt.plot(t,w)
                   
                   
    
    
```
