```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def f(t,x):
    # Entrada
    if t < 1:
        u = 0
    else:
        u = 1
    R = 1
    C = 1
    return 1/(R*C)*(u-x) 
x0 = [0]

sol = solve_ivp(f,[0,10],x0)
plt.plot(sol.t,sol.y[0])
```

```python
def f2(t,x):
    # Entrada
    if t < 1:
        u = 0
    else:
        u = 1
    R = 1
    C = 1
    L = 1
    return [x[1],-R/L*x[1]-1/L/C*x[0]+1/L/C*u] 
x0 = [0,0]

sol2 = solve_ivp(f2,[0,10],x0)
plt.plot(sol2.t,sol2.y[0])
```
