```python
import numpy as np
import matplotlib.pyplot as plt

# parametros del sistema
M = 20
B = 2
K = 5
# dy/dt=f(y,u,t)
def f(y,u,t):
    x = y[0]
    v = y[1]
    f1 = v
    f2 = -B/M*v-K/M*x+1/M*u
    return f1, f2

# Tiempo simulacion
tfin = 100
h = 0.05
t = np.arange(0,tfin,h)

# entrada
u = np.ones(len(t))

# condiciones iniciales
x0 = 0
v0 = 0

# varialbes informaticas de salida
x = np.zeros(len(t))
v = np.zeros(len(t))
x[0] = x0
v[0] = v0
for i in range(len(t)-1):
    
    k11,k12 = f([x[i],v[i]],u[i],t[i])
    xtemp = x[i] + h * k11
    vtemp = v[i] + h * k12
    k21,k22 = f([xtemp,vtemp],u[i+1],t[i+1])

    x[i + 1] = x[i] + h * (k11 + k21)/2
    v[i + 1] = v[i] + h *(k12+k22)/2

plt.plot(t,x)
```
