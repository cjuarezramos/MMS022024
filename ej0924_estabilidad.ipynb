```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig
from scipy.integrate import solve_ivp
```

```python
# parametros
g = 9.81
L = 7
m = 100
B =0.5

A = np.array([[0,1],[-g/L,-B/m]])
print(A)

eig(A)

```

```python
u = 0.1
def fnln(t,x):
    theta = x[0]
    omega = x[1]
    f1 = omega
    f2 = -B/m*omega -g/L*np.sin(theta) + u*m/L**2
    return [f1,f2]
def fln(t,x):
    theta = x[0]
    omega = x[1]
    f1 = omega
    f2 = -B/m*omega - g/L*theta + u*m/L**2
    return [f1,f2]

# condiciones iniciales
theta0 = 0
omega0 = 0
tfin = 300
sol = solve_ivp(fnln,[0,tfin],[theta0,omega0])
sol2 = solve_ivp(fln,[0,tfin],[theta0,omega0])
plt.plot(sol.t,sol.y[0])
plt.plot(sol2.t,sol2.y[0])
```
