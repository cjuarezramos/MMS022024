```python
# Autor: Carlos Juarez
# Fecha: 26-09-2024
# comentarios: Soluciona sistema de primer orden
# por medio de RK4

# Librerias
import numpy as np
import matplotlib.pyplot as plt

# Datos del sistema
R = 32
C = 0.8

# definición de funcion f
def f(t,x):
    ec = x
    # definicion de entrada
    if t < 25.6:
        v = 0
    else:
        v = 1

    return -1/(R*C)*ec + 1/R/C*v

# Datos de tiempo para la simulacion
tfin = 128
h = 2.56
n = int(tfin/h)

# arreglos para almacenar la solucion
t = np.zeros(n)
ec = np.zeros(n)

# condiciones iniciales
ec[0] = 0

# programa de Runge Kutta orden 4
for i in range(n-1):
    k1 = f(t[i],ec[i])
    ttemp = t[i]+1/2*h
    xtemp = ec[i] + 1/2*k1*h
    k2 = f(ttemp,xtemp)
    ttemp = t[i]+1/2*h
    xtemp = ec[i] + 1/2*k2*h
    k3 = f(ttemp,xtemp)
    ttemp = t[i]+h
    xtemp = ec[i] + k3*h
    k4 = f(ttemp,xtemp)
    ec[i+1] = ec[i] + 1/6*(k1+2*k2+2*k3+k4)*h
    t[i+1] = t[i]+h

plt.plot(t,ec)
```
