```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# datos del motor
R = 2.27 #â„¦
L = 0.0047 #H	
B = 0.003026 #kg.m2/s
J = 0.00246 #kg. m2	
Kb = 0.25 #V/rad/s
Km = 0.25 #N.m/A
K = 10
def f(t,x):
    i = x[0]
    w = x[1]
    if t<0.1:
        wr = 0
    else:
        wr = 43
    if t < 0.5:
        Tl = 0
    else:
        Tl = 0.5
    error = wr-w
    vin = K * error
    f1 = -R/L*i - Kb/L*w + 1/L*vin
    f2 = Km/J*i - B/J*w - 1/J*Tl
    return [f1,f2]

resultado = solve_ivp(f,[0,1],[0,0])
```

```python
plt.plot(resultado.t,resultado.y[1])
```
